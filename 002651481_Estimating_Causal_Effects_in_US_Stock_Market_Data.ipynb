#**Causality:**

This refers to the relationship between cause and effect. It's a fundamental concept that underlies how we understand the world, implying that certain events (causes) directly lead to other events (effects). Causality is a key principle in various fields, including physics, philosophy, and statistics.

#**Causal Inference:**

This is a statistical approach used to determine if a causal relationship exists between variables. It involves methods and techniques to draw conclusions about causality based on data. Causal inference is particularly important in fields like epidemiology, economics, and social sciences, where researchers are often interested in understanding the impact of policies, treatments, or other interventions.

In summary, causality is the broader concept of cause and effect, while causal inference is a set of statistical methods used to determine whether and how causality exists in specific contexts.

#**"Defining Causal Effects Using Potential Outcomes"**

We will dive deeper into the **Rubin's Causal Model (RCM)** and how it can be applied to our financial dataset.

The RCM, also known as the potential outcomes framework, is foundational in understanding causal inference.







Theory: Rubin's Causal Model (RCM)
Rubin's Causal Model conceptualizes causality by considering potential outcomes. For each unit (e.g., a company in our dataset), we imagine two potential outcomes:





 ![image.png](https://images.slideplayer.com/25/8176609/slides/slide_13.jpg)

#**The Rubin Causal Model (RCM) is based on the idea of potential outcomes.**

#**Example-**
 Enos (2014) measured attitudes toward immigration among Boston commuters. Individuals were exposed to one of two possible conditions, and then their attitudes towards immigrants were recorded.

One condition was waiting on a train platform near individuals speaking Spanish. The other was being on a train platform without Spanish-speakers. To calculate the causal effect of having Spanish-speakers nearby, we need to compare the outcome for an individual in one possible state of the world (with Spanish-speakers) to the outcome for that same individual in another state of the world (without Spanish-speakers).


In most circumstances, we are interested in comparing two experimental manipulations, one generally termed “treatment” and the other “control.” The difference between the potential outcome under treatment and the potential outcome under control is a “causal effect” or a “treatment effect.” According to the RCM, the causal effect of being on the platform with Spanish-speakers is the difference between what your attitude would have been under “treatment” (with Spanish-speakers) and under “control” (no Spanish-speakers).

The commuter survey consisted of three questions, each measuring agreement on a 1 to 5 integer scale, with 1 being liberal and 5 being conservative. For each person, the three answers were summed, generating an overall measure of attitude toward immigration which ranged from 3 (very liberal) to 15 (very conservative). If your attitude towards immigrants would have been a 13 after being exposed to Spanish-speakers and a 9 with no such exposure, then the causal effect of being on a platform with Spanish-speakers is a 4-point increase in your score.





#**We imagine two potential outcomes for our financial dataset:**

1-**What happens to the outcome of interest (like a financial metric) if the unit is exposed to a treatment (e.g., a new financial policy)?**


2-**What happens if it is not exposed?**


#**Implementation**

Suppose we want to estimate the effect of a specific financial policy or event on a company's stock price.

We'll create a hypothetical scenario where we compare companies exposed to this policy/event with those that are not, under the assumption that all other factors are constant.

#**To conduct a more complex analysis:**

We will use:

**PSM** is a statistical technique that attempts to estimate the effect of a treatment by accounting for the covariates that predict receiving the treatment. This method helps in reducing selection bias by equating groups based on these covariates.


Steps for Propensity Score Matching:

Select Covariates: Choose relevant covariates that might affect both the treatment and the outcome.

Calculate Propensity Scores: Use logistic regression to estimate the propensity score - the probability of receiving the treatment given the covariates.

Match Treated and Untreated Units: Pair each treated unit with one or more untreated units with similar propensity scores.

Analyze Matched Data: Compare the outcomes of the matched groups.

For this analysis, let's assume we choose a set of covariates that are likely to affect a company's choice of exchange and its 'EVToRevenue' ratio. These might include variables like 'Sector', 'Country', 'MarketCapitalization', etc.

```python
#Implementing Propensity Score Matching in Python ensuring that necessary libraries are installed:
!pip install pandas numpy sklearn
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# 1) Load the same CSV you used before
file_path = '/content/kaggle_finance_company_Q2_2023_overview.csv'  # or your correct path
data = pd.read_csv(file_path)

# 2) Convert columns to numeric as needed
data['MarketCapitalization'] = pd.to_numeric(data['MarketCapitalization'], errors='coerce')
data['EVToRevenue'] = pd.to_numeric(data['EVToRevenue'], errors='coerce')

# 3) Drop missing MarketCapitalization or EVToRevenue
data.dropna(subset=['MarketCapitalization', 'EVToRevenue'], inplace=True)

# 4) Create dummy variables for 'Sector' and 'Country', if they exist
if 'Sector' in data.columns and 'Country' in data.columns:
    data = pd.get_dummies(data, columns=['Sector', 'Country'], drop_first=True)

# 5) Create the Treatment variable (NASDAQ=1, other=0)
data['Treatment'] = np.where(data['Exchange'] == 'NASDAQ', 1, 0)

# 6) Select covariates for the propensity model
covariate_cols = ['MarketCapitalization'] + [col for col in data.columns if col.startswith('Sector_') or col.startswith('Country_')]
X = data[covariate_cols]
y = data['Treatment']

# 7) Standardize covariates
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 8) Fit logistic regression for propensity scores
log_reg = LogisticRegression()
log_reg.fit(X_scaled, y)
propensity_scores = log_reg.predict_proba(X_scaled)[:,1]

# 9) Store propensity scores and split into treated & untreated
data['PropensityScore'] = propensity_scores
treated   = data[data['Treatment'] == 1].copy()  # NASDAQ
untreated = data[data['Treatment'] == 0].copy()  # Others

# 10) Match each untreated unit to its nearest neighbor in treated
matched_indices = untreated['PropensityScore'].apply(
    lambda ps: np.argmin(np.abs(treated['PropensityScore'] - ps))
)
matched = treated.iloc[matched_indices]

# 11) Compute the matched effect
matched_effect = matched['EVToRevenue'].mean() - untreated['EVToRevenue'].mean()

```

```python
print("Matched Effect on EVToRevenue:", matched_effect)
```

#**Calculating the ATT score to give a more refined estimate of the causal effect by ensuring better matches between treated and control units.**

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler


# Data cleaning and preparation
data['MarketCapitalization'] = pd.to_numeric(data['MarketCapitalization'], errors='coerce')
data['EVToRevenue'] = pd.to_numeric(data['EVToRevenue'], errors='coerce')
data = data.dropna(subset=['MarketCapitalization', 'EVToRevenue'])

if 'Sector' in data.columns:
    data = pd.get_dummies(data, columns=['Sector'], drop_first=True)

# Treatment variable - assuming 'Exchange' column exists
data['Treatment'] = np.where(data['Exchange'] == 'NASDAQ', 1, 0)

# Selecting covariates
covariates = ['MarketCapitalization'] + [col for col in data.columns if 'Sector_' in col]
X = data[covariates]
y = data['Treatment']

# Standardize covariates
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Logistic regression to calculate propensity scores
log_reg = LogisticRegression()
log_reg.fit(X_scaled, y)  # Fit the model
propensity_scores = log_reg.predict_proba(X_scaled)[:, 1]

# Add propensity scores to the dataframe
data['PropensityScore'] = propensity_scores

# Separate treated and untreated groups
treated = data[data['Treatment'] == 1]
untreated = data[data['Treatment'] == 0]

# Define a caliper
caliper = 0.05 * np.std(propensity_scores)

# Implement caliper matching
matches = {}
for i in treated.index:
    pscore = treated.loc[i, 'PropensityScore']
    # Find potential matches within the caliper
    potential_matches = untreated[np.abs(untreated['PropensityScore'] - pscore) <= caliper]
    if not potential_matches.empty:
        # Select the closest match
        closest_match_index = potential_matches['PropensityScore'].sub(pscore).abs().idxmin()
        matches[i] = closest_match_index

matched_data = data.loc[matches.values()]
att = treated['EVToRevenue'].mean() - matched_data['EVToRevenue'].mean()
print(att)
```

The ATT (Average Treatment Effect on the Treated) score of 898.3112, derived from advanced propensity score matching analysis, provides a valuable insight into the causal effect of the treatment (being listed on NASDAQ) on the outcome variable (EVToRevenue) for the treated group (companies listed on NASDAQ).

### Interpretation:
- **Magnitude of the Effect**: An ATT score of 898.3112 is quite substantial. It indicates that, on average, companies listed on NASDAQ have an 'EVToRevenue' ratio that is approximately 898.3112 units higher than similar companies not listed on NASDAQ, after controlling for confounding factors using propensity score matching.

- **Indicative of a Strong Causal Relationship**: This significant difference suggests that the listing on NASDAQ might be associated with higher 'EVToRevenue' ratios. The use of propensity score matching with calipers helps ensure that the comparison is made between companies that are similar in other observed characteristics, strengthening the causal interpretation.

- **Reflection of Market Perception**: The high ATT value may reflect market perceptions that NASDAQ-listed companies, often associated with technology and growth sectors, have higher growth potential or profitability relative to their current revenues compared to companies listed on other exchanges.

### Considerations:

1. **Sector and Market Dynamics**: The result could be influenced by the type of companies that typically list on NASDAQ (like tech companies), which may inherently have different financial profiles compared to companies in more traditional industries.

2. **Practical Implication**: For stakeholders, such as investors or financial analysts, this insight could be crucial in understanding how market valuation metrics differ between exchanges and what factors might drive these differences.

In summary, the ATT score of 898.3112 suggests a significant and positive effect of being listed on NASDAQ on the 'EVToRevenue' ratio, highlighting the potential causal impact of the exchange on this financial metric.

#**Directed Acyclic Graph**

A DAG, or Directed Acyclic Graph, is a fundamental tool used in causal inference to visualize and analyze the relationships between variables. Let's break down what this means and why DAGs are important in the context of causal inference:

### What is a DAG?

1. **Directed**: This refers to the arrows in the graph, which indicate a direction of influence from one variable to another.

2. **Acyclic**: This means that the graph does not contain any cycles. In other words, you cannot start at one variable, follow the directions of the arrows, and return to the original variable.

3. **Graph**: A DAG is a graphical representation, where nodes represent variables and edges (arrows) represent direct influences or causal relationships.

### Why is a DAG Used in Causal Inference?

1. **Visualizing Relationships**: DAGs provide a clear visual way to represent and analyze the causal relationships between variables. This can help researchers and analysts identify and understand complex interdependencies.

2. **Identifying Confounders**: One of the key uses of a DAG in causal inference is to identify potential confounders. Confounders are variables that influence both the independent variable (cause) and the dependent variable (effect), potentially leading to spurious associations. DAGs help in spotting these confounders so that they can be controlled for in the analysis.

3. **Clarifying Causal Pathways**: DAGs help in distinguishing direct from indirect effects. For example, A might affect C directly or indirectly through B (A → B → C). Understanding these pathways is crucial for causal inference.

4. **Avoiding Bias**: By explicitly representing assumptions about the causal structure, DAGs can help researchers avoid biases that arise from omitting important variables or failing to consider indirect paths.

5. **Facilitating Statistical Analysis**: With the aid of a DAG, researchers can more effectively choose the appropriate statistical models and techniques to test causal hypotheses.

6. **Communicating Findings**: DAGs provide a straightforward means to communicate complex causal relationships and assumptions to others, making them a valuable tool in both research and decision-making contexts.


In summary, DAGs are a powerful tool in the realm of causal inference. They help in visualizing and analyzing causal relationships, identifying confounders, clarifying causal pathways, avoiding biases, facilitating statistical analysis, and effectively communicating findings.

#**Sensitivity Analysis**
To perform a sensitivity analysis on the propensity score matching results, we can vary some key parameters in the model, like the caliper width or the matching algorithm, and observe how the Average Treatment Effect on the Treated (ATT) changes.

This approach provides insights into the robustness of the findings against changes in model specifications.

The provided code is an implementation of a causal inference technique known as propensity score matching, specifically using nearest neighbors matching within a specified caliper width.

The code is designed to estimate the Average Treatment Effect on the Treated (ATT) and perform a sensitivity analysis by varying the caliper width.


#Objective:

- **Causal Inference Objective**: The goal here is to estimate the causal effect of the 'treatment' on the 'outcome'. In causal inference, we often want to understand what would happen to an outcome if a unit receives a treatment versus if it does not.

- **Propensity Score Matching**: This method is used to approximate a randomized control trial in observational data. By matching treated units with similar control units (based on covariates), the method attempts to balance the treated and control groups so that they are comparable.

- **ATT Estimation**: The ATT provides an estimate of the causal effect of the treatment on the outcome for those units that actually received the treatment. It's a measure of the impact of the treatment in the real world, where not everyone receives the treatment.

- **Sensitivity Analysis**: By varying the caliper width and observing changes in ATT, the sensitivity analysis assesses how robust the estimated causal effect is to the method of matching. This helps in understanding the reliability of the findings.

### Important Considerations:

- **Treatment and Outcome Variables**: In the code, 'GrossProfitTTM' and 'RevenueTTM' are used as treatment and outcome.

- **Assumptions**: Propensity score matching relies on the assumption of unconfoundedness, meaning all relevant confounders are measured and included as covariates. Any unmeasured confounding can bias the results.

In summary, this is a tool in the causal inference toolkit, specifically for estimating causal effects in observational data where randomized experiments are not feasible.

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors

# Define your columns
treatment_col = 'GrossProfitTTM'
outcome_col = 'RevenueTTM'
covariates = [col for col in data.columns if col not in [treatment_col, outcome_col]]

# Function to perform matching and calculate ATT
def calculate_att(data, caliper_width):
    treated = data[data[treatment_col] == 1]
    control = data[data[treatment_col] == 0]

    if treated.empty or control.empty:
        return np.nan  # Return NaN if either group is empty

    # Nearest neighbor matching within caliper
    nn = NearestNeighbors(n_neighbors=1, algorithm='ball_tree').fit(control[covariates])
    distances, indices = nn.kneighbors(treated[covariates], return_distance=True)

    # Apply caliper
    within_caliper = distances.flatten() <= caliper_width
    if not any(within_caliper):
        return np.nan

    matched_indices = indices.flatten()[within_caliper]

    # Calculate ATT
    matched_controls = control.iloc[matched_indices]
    att = treated.loc[within_caliper, outcome_col].mean() - matched_controls[outcome_col].mean()
    return att

# Sensitivity Analysis: Varying the caliper width
caliper_values = [0.01, 0.05, 0.1, 0.2]
att_results = {caliper: calculate_att(data, caliper) for caliper in caliper_values}

# Output the results
for caliper, att in att_results.items():
    if att is not np.nan:
        print(f'Caliper Width: {caliper}, ATT: {att}')
    else:
        print(f'Caliper Width: {caliper}, No valid matches found')

```

1. **Understanding "Valid Matches"**: A "valid match" refers to a pair of treated and control units whose propensity scores are within the specified caliper distance.

  The smaller the caliper width, the closer in propensity score (and thus, presumably, in covariates) the matched units are.

2. **Interpretation of Match Counts**:

   - **Caliper Width 0.01**: At this very narrow caliper width, we're finding between 1 to 4 valid matches for different treated units.
   
    This indicates that only a few control units are closely similar to the treated units in terms of propensity scores.

   - **Increasing Caliper Widths (0.05, 0.1, 0.2)**: As the caliper width increases, the number of matches found remains relatively consistent.
   
    This suggests that even with a broader caliper, the number of closely similar control units does not increase significantly.

In summary, these results provide insights into the feasibility and robustness of matching treated and control units in the dataset.
